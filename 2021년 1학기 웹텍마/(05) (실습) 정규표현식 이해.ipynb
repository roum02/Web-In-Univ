## Part 2 - sub & split

```python
# (ì‹¤ìŠµ) 
# ë¬¸ìì—´ì—ì„œ íŠ¹ì • ë‹¨ì–´ë¥¼ ìˆ˜ì •í•˜ê¸° (1) 

text5 = "apple box orange tree"

change_text = re.sub('apple|orange', 'fruit' , text5)   # ë¬¸ìì—´ì— ìˆëŠ” apple ê³¼ orange ë¥¼ fruit ë¡œ ìˆ˜ì •
#ì• í”Œì´ë‚˜ ì˜¤ëœì§€ë¥¼ ì°¾ì•„ì„œ í”„ë£¨ì¸ ë¡œ ë°”ê¾¸ì–´ì¤„ ê²ƒ

print(change_text)     # ìˆ˜ì •ëœ ë¬¸ìì—´ì„ ì¶œë ¥ 
```

```python
# (ì‹¤ìŠµ) 
# ë¬¸ìì—´ì—ì„œ íŠ¹ì • ë‹¨ì–´ë¥¼ ìˆ˜ì •í•˜ê¸° (2) 

text6 = "1 3 apple box 5 orange tree 7 9"


change_char = re.sub('[0-9]+' , 'number', text6)    # ìˆ«ìë§Œ ì°¾ì•„ì„œ number ë¼ëŠ” ë‹¨ì–´ë¡œ ë°”ê¾¸ì–´ ê²°ê³¼ ë¬¸ìì—´ì„ ë°˜í™˜
#ìˆ«ìë¥¼ ì°¾ì•„ì„œ ë„˜ë²„ë¡œ ë°”ê¿”!
#+ëŠ” 0-9ê¹Œì§€ í•œ ìë¦¬ ì´ìƒì„ ì°¾ì•„ì£¼ì„¸ìš”
print(change_char)     # change_char ëŠ” ë¬¸ìì—´ ì„ 
```

```python
# ë§¤ê°œë³€ìˆ˜ë¡œ í•¨ìˆ˜ ì´ìš©í•˜ê¸° 
# ìœ„ ì…€ì—ì„œ text = "1 3 apple box 5 orange tree 7 9"

def multiple10(m):          # ë§¤ê°œë³€ìˆ˜ë¡œ ë§¤ì¹˜ ê°ì²´ë¥¼ ë°›ìŒ
    n = int(m.group())      # ë§¤ì¹­ëœ ë¬¸ìì—´ì„ ê°€ì ¸ì™€ì„œ ì •ìˆ˜ë¡œ ë³€í™˜
    return str(n * 10)
#mì„ ì°¾ì•„ì„œ 10ì„ ê³±í•´ì„œ ë¬¸ìë¡œ ë°”ê¿ˆ

change_char = re.sub('[0-9]+' ,multiple10  , text6)    # ìˆ«ìë§Œ ì°¾ì•„ì„œ 10ì„ ê³±í•´ì¤€ í›„ ê²°ê³¼ ë¬¸ìì—´ì„ ë°˜í™˜

print(change_char)
```

```python
# (ì‹¤ìŠµ) 
# ë¬¸ìì—´ì—ì„œ íŒê¸€ë§Œ ì¶”ì¶œí•˜ê¸° 
# ì¶”ì²œí•˜ëŠ” ë°©ë²•


text7 = "ì¸ê³µì–´(äººå·¥èª): (ì‚¬íˆ¬ë¦¬ ë”°ìœ„ ìì—°ì–´(è‡ªç„¶èª)ì— ëŒ€(å°)í•˜ì—¬) ì˜ì‹ì (æ„è­˜çš„)ìœ¼ë¡œ ì°½ì¡°(å‰µé€ )ëœ ì–¸ì–´(è¨€èª)"

reg = re.compile("[^ê°€-í£ ]")    # í•œê¸€ì´ ì•„ë‹Œ ë¬¸ìë§Œ ë§¤ì¹˜í•˜ëŠ” ì •ê·œì‹ - 'í£' ë¬¸ì ë’¤ì— blank ê°€ ìˆì–´ì•¼ ë‹¨ì–´ í˜•ì‹ìœ¼ë¡œ ì¶œë ¥ !!! 
#regëŠ” í•œê¸€ì´ ì•„ë‹Œ ê²ƒë§Œ ì°¾ì•„ì£¼ì„¸ìš” -> ì´ ê²½ìš° from ìƒëµ ê°€ëŠ¥
#ë§¨ ë’¤ ë°˜ë“œì‹œ í•œ ì¹¸ì„ ë„ì–´ì•¼ í•©ë‹ˆë‹¤

reg.sub('', text7)     # í•œê¸€ì´ ì•„ë‹Œ ë¬¸ìëŠ” ëª¨ë‘ ì œê±°í•œ í›„ ê²°ê³¼ ë¬¸ìì—´ì„ ë°˜í™˜
#sub ëŒ€ì²´í•˜ëŠ” ê²ƒ
```

```python
# (ì‹¤ìŠµ)  
# split ë©”ì†Œë“œ ì‚¬ìš©ë²• - ê²°ê³¼ ë¬¸ìì—´ì„ ë°˜í™˜ 

text8 = "hi! hello world! hello python! hello!! hello!?"   

reg = re.compile(r"\s*hello\s*" )    # ë¬¸ìì—´ì—ì„œ hello ë¥¼ ì°¾ëŠ” ì •ê·œì‹ 

print(reg.split(text8))   # hello ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë¬¸ìì—´ì„ ë‚˜ëˆ„ì–´ ê²°ê³¼ë¥¼ ë°˜í™˜ (hello ë‹¨ì–´ëŠ” ë¯¸í¬í•¨)
#helloë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë¬¸ìì—´ì„ ë¶„ë¦¬í•´ì¤˜!
```
