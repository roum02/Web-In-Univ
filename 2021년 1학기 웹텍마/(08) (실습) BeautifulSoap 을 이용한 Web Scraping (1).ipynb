```python
# 필요한 package 를 import


import requests
from bs4 import BeautifulSoup #(as bs)   #패키지 이름 시작은 대문자임(B,S)


import warnings
warnings.filterwarnings('ignore')
```

## ▶Requests class 와 BeautifulSoap class 사용 예 

### - (e.g.) 
https://en.wikipedia.org/wiki/Natural_language_processing


```python
# 위 url 을 클락하여 크롬 브라우저에서 "natural language processing" 을 검색 
# 검색 후 브라우저에서 url 을 복사하여 아래 해당 변수에 입력  
# requests.text : 해당 URL 의 내용을 readable한 내용으로 디코딩 
# requests.content: Binary 형식으로 반환, 이미지 등의 미디어 파일을 받고자 할 때 사용

url = "https://en.wikipedia.org/wiki/Natural_language_processing" 

nlp = requests.get(url)    # 해당 페이지를 가져옴

nlp.text       # 해당 페이지의 내용을 확인 
```

```python
# BeautifulSoup 을 이용하여 내용을 파싱 (html 을 해석할 수 있는 아이)

soup = BeautifulSoup(nlp.text) #얘가 정리해줌

soup     # 파싱 후 해당 페이지의 내용을 확인
```

## ▶ beautifulsoup 메소드들 (3)

### - 해당 페이지에서 태그 내의 내용을 가져오고자 할 때

```python
#  아래의 경우는 첫번째 div tag 내용만 가져옴.  해당 페이지에는 많은 div tage가 존재

soup.find("div")
 
```

```python
# find 메소드와 같은 기능을 수행

soup.select_one("div")
```

```python
# find 와 select 의 차이 
# head 태그 아래있는 title태그를 찾고자 할 때 
# find 의 경우 


soup.find("head").title

#계속 .을 찍고 계속 적어주어야 함.
```

```python
# find 와 select 의 차이 
# head 태그 아래있는 title태그를 찾고자 할 때 
# select 의 경우 

soup.select_one("head > title")   # path 를 지정해 주어야 함 
```

```python
# attrs 매개변수를 이용하여 추가적인 정보를 지정할 수 있음 
# div 의 id 를 지정해야....
# id 속성은 전체 문서에서 유일함 

soup.find("div", attrs={"id":"toc"})
```

```python
# text 만을 출력하고자 할 때

print(soup.find("div", attrs={"id":"toc"}).text)
```

```python
# 페이지내 동일한 태그가 다수 존재 할 때 find_all 이용하여 검색 가능

soup.find_all('b')
```

```python
# select 메소드는 find_all 과 같은 기능을 수행


soup.select('b')
```

```python
# 페이지 내에서 link 가 연결된 태그를 검색하고자 할 때

soup.find_all('a')

#a 태그가 다 나옴
```

```python
# 여러 elements 중에서 특정 element 를 지정헤서 가져 올 때 


print(soup.find_all('a')[10].get_text()) #10번째 a태그
print()

print(soup.find_all('a')[15].get_text()) #15번째 a태그

```

```python
# 위 내용 중에서 href 의 내용만을 추출하고자 할 때


for contents in soup.find_all('a'):
    if 'href' in contents.attrs :     #속성 중에 'href'가 있다면,출력해주세요.
        print(contents.attrs['href'])  # 출력 내용 중 #으로 시작되는 것은 페이지내로 이동하는 경우를 말함 

#뉴스 페이지의 해드만 가져오기.. 등등에 사용됨
```

```python
# get 메소드를 이용하여 href 의 내용만을 추출할 수도 있음 
# 
# 위와 동일 결과를 얻을 수 있음 

for contents in soup.find_all('a'):
    print(contents.get('href'))
    
#가장 첫 a태그에 href내용이 없어서 None으로 나옴    
```

```python
# 페이지 내 이동이 아닌, 순수 다른 페이지로 이동하는 경우만을 추출할 때


for contents in soup.find_all('a'):
    if 'href' in contents.attrs :
        if '#' not in contents.attrs['href'] :  # \#이 보기 싫어서 뺌
            print(contents.attrs['href'])
            
```
