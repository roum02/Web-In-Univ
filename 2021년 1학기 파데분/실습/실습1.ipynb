```python
#지하철 자동 발권 

money = int(input("투입한 돈은?"))
first_section = int(input("1구간 표의 수는?"))
second_section = int(input("2구간 표의 수는?"))

change = money - ((700*first_section)+(900*second_section))
print("잔돈",change)

ten_thousand = int(change/1000)
print("1000원 지폐:",ten_thousand,"장")
one_hundred = int((change - (ten_thousand*1000))/100)
print("100원 동전:", one_hundred,"개")
```
