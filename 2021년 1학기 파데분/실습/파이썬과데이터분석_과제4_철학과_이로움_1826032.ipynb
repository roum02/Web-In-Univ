**[과제 4] Tic-Tac-Toe 게임 구현하기**

<구현에 필요한 함수> 

*  make_board: board로 사용할 2차원 리스트 생성 
  *  parameter: 없음 
*  display_board: board의 현재 상태를 출력 
  *  parameter: 없음 
*  check_win: 새로운 mark에 의해 승자가 결정되었는지 판단
  *  parameter:mark 현재 player의 심볼(X or O)을 저장 
*  play: 현재 player로 부터 mark할 위치를 입력 받아서 그 위치에 mark함 
  *  parameter: mark 
    *  현재 player의 심볼(X or O)을 저장


<조건 1> 게임을 시작하면 make_board 함수를 사용해서 board를 만들고 board의 상태를 출력한다. 

<조건 2> play함수는 player가 입력 값이 board의 범위를 벗어나지 않은지 확인하고 범위에 맞는 값을 입력할 때까지 입력을 받는다 

<조건 3> player가 선택을 끝내면 board를 상태를 출력한다 

<조건 4> 게임은 승자가 결정되거나 더 이상 선택할 칸이 없으면 종료한다. 

<조건 5> 게임 종료시 승자 또는 무승부를 알려준다. (실행 결과 참조) 

<조건 6> 시작하는 player는 'X' mark를 사용하고 player는 번갈아 게임에 참가한다.


```python
inner_list = []
outter_list = []

#board로 사용할 2차원 리스트를 생성하는 함수
def make_board():
  global inner_list
  global outter_list

  #2차원 리스트 만들기
  for i in range(0, 3):
    for k in range(0, 3):
      inner_list.append(' ')
    outter_list.append(inner_list)
    inner_list = []
  print("게임판 생성")
  print("")

#board의 현재 상태를 출력하는 함수
def display_board():
  #global outter_list
  for i in range(0, 3):
    print(outter_list[i])
  print("")
  print("========================")
  print("")

#새로운 mark에 의해 승자가 결정되었는지 판단.
def check_win(mark):
    return (outter_list[0][0]+outter_list[0][1]+outter_list[0][2] == mark *3) or \
    (outter_list[1][0]+outter_list[1][1]+outter_list[1][2] == mark *3) or \
    (outter_list[2][0]+outter_list[2][1]+outter_list[2][2] == mark *3) or \
    (outter_list[0][0]+outter_list[1][0]+outter_list[2][0] == mark *3) or \
    (outter_list[0][1]+outter_list[1][1]+outter_list[2][1] == mark *3) or \
    (outter_list[0][2]+outter_list[1][2]+outter_list[2][2] == mark *3) or \
    (outter_list[0][0]+outter_list[1][1]+outter_list[2][2] == mark *3) or \
    (outter_list[2][0]+outter_list[1][1]+outter_list[0][2] == mark *3)


#현재 player로부터 mark할 위치를 입력 받아서 그 위치에 mark함
def play(mark):
  
  #입력 값이 범위에 맞을 때까지 반복
  while True:
    #행과 열 값을 입력받음
    row, column = input("행렬의 행과 열을 입력하시오 (예)0, 1 :").split(",")
  
    #정수값으로 바꾸어준다.
    row = int(row)
    column = int(column)
    
    #입력 값이 board의 범위를 벗어나지 않은지 확인
    if row < 3 and column < 3:
      #입력값이 이미 표기되어 있지 않을 경우
      if outter_list[row][column] != ' ':
        print("이미 표기한 칸입니다. 다시 선택하세요.")
        print("")
      else:
        break
    else:
      print("범위를 벗어났습니다. 0부터 2사이의 정수값을 입력하세요.")
      print("")

  #해당 위치에 mark
  outter_list[row][column] = mark

#실제 게임을 실행할 함수 생성
def game_start():
  #X player와 O player 순서 구분을 위한 변수
  play_number = 1
  
  global inner_list
  global outter_list

  #리스트 초기화
  inner_list = []
  outter_list = []

  make_board()
  display_board()
  print("X player 부터 게임을 시작합니다.")
  print("")

  while True:
    #홀수이면 X의 차례
    if play_number%2 != 0:
      mark = "X"
      play(mark)
      print("")
      display_board()
      play_number += 1
      if check_win(mark) == True:
        print("X player가 승리했습니다.")
        break
      elif play_number > 9 and check_win(mark) == False:
        print("무승부 입니다.")
        break
    #짝수라면 O의 차례
    else:
      mark = "O"
      play(mark)
      print("")
      display_board()
      play_number += 1
      if check_win(mark) == True:
        print("O player가 승리했습니다.")
        break 
```

<실행 결과1>

```python
game_start()
```
