```python
#무한루프를 활용해 입력한 두 숫자의 합계를 반복하여 계산

first_num = 0
last_num = 0
sum = 0

while True:
  first_num = int(input("더할 첫 번째 수를 입력하세요"))
  last_num = int(input("더할 두 번째 수를 입력하세요"))
  sum = first_num + last_num
  print(sum)
```

```python
#사용자가 멈출때까지 연산자까지 계산

first_num = 0
last_num = 0
char = ""

while True:
  first_num = int(input("연산할 첫 번째 수를 입력하세요"))
  last_num = int(input("연산할 두 번째 수를 입력하세요"))
  char = input("계산할 연산자를 입력하세요")

  if char == "+":
    print("%d + %d = %d" %(first_num, last_num, first_num + last_num))
  elif char == "-":
    print("%d - %d = %d" %(first_num, last_num, first_num - last_num))
  elif char == "*":
    print("%d * %d = %d" %(first_num, last_num, first_num * last_num))
  elif char == "/":
    print("%d / %d = %d" %(first_num, last_num, first_num / last_num))
```

```python
#마름모 출력하기 while

i,k = 0,0

while i < 9:
  if i < 5:
    while k < 4-i:
      print(" ", end="")
      k += 1
    k = 0
    while k <(2*i)+1:
      print("e",end='')
      k += 1
    k = 0
  else:
    while k <i-4:
      print(" ", end='')
      k += 1
    k = 0
    while k <(9-i)*2 -1:
      print("e",end='')
      k += 1
    k = 0

  i += 1
  print()
```
